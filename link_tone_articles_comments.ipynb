```r
library(readr)
library(MASS)
library(dplyr)
```

```r

#we read the dataframe for articles and comments 
articles <- read_csv('final_df_without_commentsentiment.csv')

comments <- read_csv('comments_with_lexicoder_sentiment.csv')
```

```r
#we first need to group comments by article and calculate average, max, and min negativity scores
comments_agg <- comments %>%
  group_by(article_url) %>%
  summarise(
    avg_negativity_comment = mean(negativity_percent, na.rm = TRUE),
    max_negativity_comment = max(negativity_percent, na.rm = TRUE),
    min_negativity_comment = min(negativity_percent, na.rm = TRUE)
  )
```

```r
#then we add the comment negativity stats to the articles dataset 
articles <- articles %>%
  left_join(comments_agg, by = "article_url")

```
